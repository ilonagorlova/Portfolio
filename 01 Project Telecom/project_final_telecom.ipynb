# Поиск клиентов с неоптимальными тарифами

## Цель проекта

В этом проекте анализируются данные о клиентах и их звонках провайдера виртуальной телефонии "Нупозвони" с целью найти пользователей, которые переплачивают на текущих тарифах и сэкономят, если перейдут на другой тариф.

Результаты этого исследования позволят сделать предложение некоторым клиентам перейти на более выгодный для них тариф и удержать их от ухода.

Исследование проводилось в начале декабря 2019 г.

**Шаг 1. Загрузка данных из предоставленных файлов.**

* анализ и обработка пропусков (при необходимости)
* анализ и обработка дубликатов (при необходимости)
* проверка соответствия названия столбцов их содержанию и змеиному регистру
* проверка соответствия типов столбцов содержимому и при необходимости изменить
* создание объединенного датафрейма, который содержит информацию о клиентах, типе тарифа и звонках
* создать дополнительных столбцы в объединенном датафрейме, которые потребуется в дальнейшем анализе

**Шаг 3. Исследовательский анализ данных:**

* распределение клиентов по тарифам
* распределение количества операторов у клиентов

* составить таблицу с клиентами, которым стоит сменить тариф (id клиента, текущий и новый оптимальный тариф)
* провести анализ экономических показателей до и после смены тарифа выбранными клиентами:
    - рассчитать выручку заказчика (общую за период исследования, за каждый месяц, от каждого клиента) до и после смены тарифа у выбранных клиентов
    - проанализировать, за какой срок окупится падения выручки за счет смены на тарифов на более оптимальные
    - проанализировать, как изменится распределение пользователей по тарифам (в случае смены тарифа)

Для подсчета числа минут входящих и исходящих (внутри и вне сети) звонков будем использовать функции.

```python
def count_min_in(row):
    """ Функция count_min_in() предназначена для подсчета количества минут входящих звонков (direction='in'),
    в качестве аргумента функция принимает строку датафрейма."""
    if row['direction'] == 'in':
        return row['total_call_duration']
    else:
        return float("nan")
```

```python
def count_min_out_inside(row):
    """ Функция count_min_out_inside() предназначена для подсчета количества минут исходящих звонков (direction='out')
    внутри сети (internal=True), в качестве аргумента функция принимает строку датафрейма."""
    if row['internal'] == True and row['direction'] == 'out':
        return row['total_call_duration']
    else:
        return float("nan")
```

```python
def count_min_out_outside(row):
    """ Функция count_min_out_outside() предназначена для подсчета количества минут исходящих звонков (direction='out')
    вне сети (internal=False), в качестве аргумента функция принимает строку датафрейма."""
    if row['internal'] == False and row['direction'] == 'out':
        return row['total_call_duration']
    else:
        return float("nan")
```

```python
df_telecom_clients.dtypes
```

```python
df_telecom_clients['in_calls_min'] = df_telecom_clients.apply(count_min_in, axis=1)
```

```python
df_telecom_clients['out_calls_inside_min'] = df_telecom_clients.apply(count_min_out_inside, axis=1)
```

```python
df_telecom_clients['out_calls_outside_min'] = df_telecom_clients.apply(count_min_out_outside, axis=1)
```

### Агрегация данных клиентов за месяц

Создадим датафрейм `monthly_data` с суммарным количеством минут входящих, исходящих звонков внутри и вне сети, а также числом уникальных операторов по месяцам для каждого клиента.

### Распределение количества мин входящих и исходящих для каждого клиента в зависимости от тарифа

```python
mean_tariff = monthly_data.pivot_table(index='tariff_plan', values=['out_calls_inside_min',
                                                                    'out_calls_outside_min', 'in_calls_min'], 
                        aggfunc='mean')
mean_tariff
```

```python
mean_tariff = mean_tariff.transpose()
mean_tariff
```

### Рассчет месячных затрат на связь у клиентов с текущим и альтернативными тарифами.

Для расчета стоимости услуг за месяц для каждого клиента будем использовать следующую формулу:
    
`Сумма услуг связи за месяц = абонентская плата за месяц + число работающих операторов за месяц * ежемесячная плата за каждого оператора + число минут исходящих звонков внутри сети сверх лимита * цена минуты исходящих звонков сверх лимита + число минут исходящих звонков вне сети * цена минуты исходящих звонков вне сети`

Для расчета стоимости услуг связи за месяц будем использовать функции:
* `get_current_costs()`- для расчета стоимости связи за месяц при текущем тарифе
* `get_monthly_cost()` - для расчета стоимости связи за месяц для каждого тарифа А,В,С

```python
def get_current_costs(row):
    """ Функция для расчёта стоимости услуг связи для текущего тарифа клиента.
В качестве аргумента использует строку датафрейма, в котором содержится
информация о сумме мин входящих, исходящих (вне и внутри сети) звонков, количестве проработавших
операторов за месяц"""
    if row['out_calls_outside_min'] <= df_tariffs.at[row['tariff_plan'],'minutes_in_tariff']:
        costs = (df_tariffs.at[row['tariff_plan'], 'monthly_pay']
             + df_tariffs.at[row['tariff_plan'],'monthly_pay_oper'] * row['operator_id']
             + 0
             + df_tariffs.at[row['tariff_plan'],'out_calls_outside'] * row['out_calls_outside_min'])
        #print(row['out_calls_outside'])
        return costs
    else:
        costs = (df_tariffs.at[row['tariff_plan'],'monthly_pay']
             + df_tariffs.at[row['tariff_plan'],'monthly_pay_oper'] * row['operator_id']
             + (row['out_calls_inside_min'] -  df_tariffs.at[row['tariff_plan'],'minutes_in_tariff'])
             * df_tariffs.at[row['tariff_plan'],'out_calls_inside']
             +  df_tariffs.at[row['tariff_plan'],'out_calls_outside'] * row['out_calls_outside_min'])    
        return costs
```

```python
def get_monthly_costs(row, tariff_type):
    """ Функция для расчёта стоимости услуг связи для тарифов А,В,С.
В качестве аргумента принимает строку датафрема (row), в котором содержится
информация о сумме мин входящих, исходящих (вне и внутри сети) звонков, количестве проработавших
операторов за месяц, а также тип тарифа (tariff_type).
Для расчета используются информация о ценах на тарифы из датафрейма df_tariffs."""
    if row['out_calls_outside_min'] <= df_tariffs.at[tariff_type,'minutes_in_tariff']:
        costs = (df_tariffs.at[tariff_type, 'monthly_pay']
             + df_tariffs.at[tariff_type,'monthly_pay_oper'] * row['operator_id']
             + 0
             + df_tariffs.at[tariff_type,'out_calls_outside'] * row['out_calls_outside_min'])
        #print(row['out_calls_outside'])
        return costs
    else:
        costs = (df_tariffs.at[tariff_type,'monthly_pay']
             + df_tariffs.at[tariff_type,'monthly_pay_oper'] * row['operator_id']
             + (row['out_calls_inside_min'] -  df_tariffs.at[tariff_type,'minutes_in_tariff'])
             * df_tariffs.at[tariff_type,'out_calls_inside']
             +  df_tariffs.at[tariff_type,'out_calls_outside'] * row['out_calls_outside_min'])    
        return costs
```

Рассчитаем стоимость услуг за каждый месяц для каждого клиента с использованием текущего тарифа, результаты запишем в новый столбец `costs_current`. 

```python
 monthly_data['costs_current'] =  monthly_data.apply(get_current_costs, axis=1)
```

Рассчитаем стоимость услуг за каждый месяц для каждого клиента с использованием тарифов A,B,C, результаты запишем в новые столбцы `costs_A`, `costs_B`, `costs_C`. 

```python
for type in df_tariffs.index:
    monthly_data[f'costs_{type}'] = monthly_data.apply(get_monthly_costs, tariff_type=type, axis=1)
```

### Выбор оптимального тарифа в.2.

Создадим датафрейм monthly_data_sum, который будет содержать суммарные значения стоимости услуг каждого клиента с разными тарифами. Для этого сгруппируем датафрейм monthly_data2 по user_id, и просуммируем нужные нам столбцы.

```python
monthly_data_sum = monthly_data.groupby('user_id').agg({'tariff_plan':'first', 'costs_current':'sum', 'costs_A':'sum',
                                                          'costs_B':'sum', 'costs_C':'sum', 'month':'count'})
monthly_data_sum
```

Выберем оптимальную стоимость тарифа - минимум из данных по столбцам `costs_A`, `costs_B`, `costs_C`.

```python
monthly_data_sum['costs_best_tariff'] = monthly_data_sum[['costs_A', 'costs_B', 'costs_C']].min(axis=1)
monthly_data_sum.head(2)
```

С помощью функции `type_tariff()` выведем тип тарифа, который соответствует оптимальной стоимости услуг связи, запишем данные в новый столбец `best_tariff_plan`.

```python
def type_tariff(row):
    """Функция type_tariff() определяет тип тарифа с наилучшей стоимостью"""
    if row['costs_best_tariff'] == row['costs_A']:
        return 'A'
    elif row['costs_best_tariff'] == row['costs_B']:
        return 'B'
    elif row['costs_best_tariff'] == row['costs_C']:
        return 'C'
```

```python
monthly_data_sum['best_tariff_plan'] =  monthly_data_sum.apply(type_tariff, axis=1)
monthly_data_sum.head(2)
```

Посчитаем разницу между стоимостью услуг связи на текущем и оптимальном тарифах в р. и в % (по отношению в текущему тарифу). Запишем получившиеся данные в столбцы `diff_current_best` и `diff_current_best_%` соответственно. 

```python
monthly_data_sum['diff_current_best'] = monthly_data_sum['costs_current'] -  monthly_data_sum['costs_best_tariff']
```

```python
monthly_data_sum['diff_current_best_%'] = (monthly_data_sum['costs_current'] 
                                        -  monthly_data_sum['costs_best_tariff'])/monthly_data_sum['costs_current']*100
```

С использованием функции `change_tariff()` создадим новый столбец `costs_after_change` в датафрейме `monthly_data_sum`, в котором будем записана стоимость оптимального тарифа (если разница в стоимости между текущим и оптимальным составляет 5 и более % ), либо стоимость текущего тарифа (если разница менее 5 %).

```python
def change_tariff(row):
    if row['diff_current_best_%'] >= 5:
        return row['costs_best_tariff']
    else:
        return row['costs_current']
```

## Гипотезы.

 ### Гипотеза о разной стоимости услуг на тарифных платах А, В, С.

Проверим гипотезы о различии стоимости услуг для тарифов А, B, C попарно (для тарифов А и B,  B и C, А и C).

**Нулевая гипотеза H0:** средняя стоимость услуг за месяц на тарифных планах А и В одинаковая.

**Альтернативная гипотеза H1:** средняя стоимость услуг за месяц на тарифных планах А и В различается.


Для расчёта статистической значимости различий в среднем числе заказов будем использовать критерий Манна-Уитни, так как согласно построенными в разделе 6.1 графикам, данные содержат сильные выбросы.
Среднее число заказов будем рассчитывать как среднее по выборкам `sampleA` и `sampleB` (так как число элементов выборки - это число всех пользователей (в том числе с нулевыми заказами), сумма элементов - количество заказов).

```python
# первая выборка - стоимость услуг связи за месяц клиентов с тарифом А
sample1 = monthly_data['costs_A']
# вторая выборка - стоимость услуг связи за месяц клиентов с тарифом В
sample2 = monthly_data['costs_B']

# выведем средние значния 2 выборок
print(f'Средняя стоимость услуг связи за месяц клиентов с тарифом А: {np.round(np.mean(sample1),2)}')
print(f'Средняя стоимость услуг связи за месяц клиентов с тарифом В: {np.round(np.mean(sample2),2)}')

# установим уровень значимости 5 %
alpha = 0.05

results = st.mannwhitneyu(sample1, sample2, alternative='two-sided')
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')    
```

```python
# составляем выборки
# первая выборка - длительность исходящих звонков внутри сети клиентов с тарифом А
sample1 = monthly_data[monthly_data['tariff_plan'] == 'A']['out_calls_inside_min']
# вторая выборка - длительность исходящих звонков внутри сети клиентов с тарифом B
sample2 = monthly_data[monthly_data['tariff_plan'] == 'B']['out_calls_inside_min']

# выведем средние значния 2 выборок
print(f'Средняя длительность исходящих звонков внутри сети за месяц у клиентов с тарифом А: {np.round(np.mean(sample1),2)} мин')
print(f'Средняя длительность исходящих звонков внутри сети за месяц у клиентов с тарифом B: {np.round(np.mean(sample2),2)} мин')

# установим уровень значимости 5 %
alpha = 0.05

results = st.mannwhitneyu(sample1, sample2, alternative='greater')
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')   
```

Нулевую гипотезу о равенстве длительности исходящих звонков внутри сети за месяц у клиентов с тарифами А и B **отвергаем** в пользу альтернативной, о том, что длительность исходящих звонков внутри сети за месяц у клиентов с тарифом А больше, чем у клиентов с тарифом B.
