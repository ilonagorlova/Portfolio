## Цели и задачи проекта

**Целью проекта** является выявление факторов успешности выпущенных компьютерных игр на основе исторических данных о продажах, оценках пользователей и экспертов, жанрах и платформах для определения потенциально популярных продуктов в 2017 г. и планирования рекламных компаний.

Заказчик - интернет-магазин "Стримчик", который продает по всему миру компьютерные игры.

В ходе исследования необходимо решить следующие **задачи**:

* загрузить данные из предоставленного файла и провести предобработку данных: проверить типы данных, обработать пропуски и дубликаты, добавить необходимые для дальнейшего анализа столбцы.
* провести исследовательский анализ следующих параметров:
    - количество выпущенных игр по годам
    - суммарные продажи по годам для разных игровых платформ
    - определить среднее жизни игровых платформ
    - выявить актуальный период для дальнейшего исследования (с целью прогноза на 2017 г.)
    - выбрать потенциально прибыльные платформы за актуальный период
    - проанализировать глобальные продажи по разным платформам
    - оценить влияние отзывов пользователей и критиков на продажи игр
    - выявить жанры с высокими и низкими продажами       
* описать "портрет" типичного пользователя игр для каждого региона (Северная Америка, Европа, Япония), а именно:
    - определеить самые популярные платформы (топ-5)
    - определить самые популярные жанры (топ-5)
    - оценить влияния возрастного рейтинга ESRB на прождажи игр в отдельных регионах

* проверить гипотезы о равенстве пользовательских рейтингов для отдельных платформ и жанров.

### Изменение названия столбцов

### Обработка пропусков

Посмотрим на количество пропусков в каждом столбце датафрейма.

```python
# выведем число пропусков в каждом столбце датафрейма
data.isna().sum()
```

<a id='Пропуски'></a>
Пропуски обнаружены в столбцах `name, year_of_release, genre, critic_score, user_score, rating`. Большое количество пропусков может быть связано с тем, что файл составлялся по данным из открытых источников и не все столбцы могли быть заполнены. Также могли возникнуть технологические ошибки при формировании файла.
Пропуски в столбце rating связаны с тем, что организация ESBR определяет рейтинг только для игр, выпущенных для Северной Америки. 

```python
# посмотрим на пропуски в столбце name
data[data['name'].isna()]
```

В столбцах `name` и `genre` встречаются по 2 пропуска в одних и тех же строках. Их можно просто удалить.

```python
# удалим строки с пропусками в столбцах name, genre 
data.dropna(subset=['name','genre'], inplace=True)
# перезапишем индексы
data.reset_index(drop=True)
```

```python
#выведем число пропуской в столбце year_of_release
data['year_of_release'].isna().sum()
```

296 пропусков в столбце из 16713 строк - это тоже не так много. Достоверно заменить год выпуска сложно, поэтому удалим эти значения тоже.

```python
# удалим строки с пропусками в столбце year_of_release
data.dropna(subset=['year_of_release'], inplace=True)
# перезапишем индексы
data.reset_index(drop=True)
# проверим количество пропусков после удаления
data['year_of_release'].isna().sum()
```

```python
len(data)
```

Выведем количество пропусков и их долю в столбцах `user_score` и `critic_score`.

```python
print('Пропуски в user_score:', data['user_score'].isna().sum(), (data['user_score'].isna().sum()/len(data)*100).round(0), '%')
print('Пропуски в critic_score:', data['critic_score'].isna().sum(), (data['critic_score'].isna().sum()/len(data)*100).round(0), '%')
```

Пропусков в столбцах `user_score` и `critic_score` много, они составляют до 51 % от общего количества строк. Удалить такое количество строк мы не можем, заполнить достоверно (средним, медианой и т.п.) их не представляется возможным, так как это повлияет на будущие оценки. Оставим эти столбцы с пропусками.
Проверим уникальные значения в столбце `user_score`. Так как у этого столбца тип object, то в нем могут встречаться символьные значения.

```python
print('Оценки пользователей, которые встречаются в датафрейме')
print(data['user_score'].sort_values().unique())
```

В таблице указана средняя оценка пользователей в диапазоне от 0 до 10. Также встречается значение 'tbd' - аббревиатура словосочетания «To Be Determined», что означает «будет определено». Эти значения могли записать в ячейку, в случае, когда  пользователи еще не успели оценить игру.
Посмотрим, сколько значений tbd встречается в столбце `user_score`.

```python
# длина среза датафрейма, в котором user_score = 'tbd'
len(data[data['user_score'] == 'tbd'])
```

В столбце `user_score` встречается 2376 значений 'tbd'. Так как по своей сути 'tbd' - это отсутствующие значения, то заменим их на NaN.

```python
# заменяем значения tbd на nan в столбце user_score
data['user_score'] = data['user_score'].replace({'tbd': np.nan})
# число nan после замены tbd
print('Пропуски в user_score:', data['user_score'].isna().sum(), (data['user_score'].isna().sum()/len(data)*100).round(0), '%')
```

После замены значений 'tdb' на Nan,  количество пропусков в столбце увеличилось до 55 %.

```python
# уникальные значения в столбце
data.user_score.unique()
```

Посмотрим на пропуски в столбце `rating`.

```python
print('Пропуски в rating:', data['rating'].isna().sum(), (data['rating'].isna().sum()/len(data)*100).round(0), '%')
```

Выведем список уникальных значений возрастных категорий, которые есть в этом столбце.

```python
 data['rating'].unique()
```

```python
# заполним значения NaN строкой 'unknown'
data['rating'] = data['rating'].fillna('unknown')
data['rating'].isna().sum()
```

Рассчитаем медианное время жизни как разницу между максимальным и минимальным годом выпуска игр для каждой платформы + 1 (включаем граничные года диапазона).
Сначала рассчитаем медианное время жизни по данным для всех платформ (31).

```python
# Комментарий ревьюера
temp = data.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # в срок жизни платформы, должны входить границы диапазона, 
                                                            # поэтому +1
time_life['life'].median()
```

**Медианное время жизни платформы**, рассчитанное по всем платформам, составляет **7 лет**.
Посмотрим, изменится ли этот параметр при расчёте для топ-3, топ-5, топ-10, топ-20 платформ в рейтинге по глобальным продажам.

```python
# создадим списки с названиями топ-3, топ-5, топ-10, топ-20 платформ по глобальным продажам и с названиями всех платформ
top3_platform = list(data.groupby('platform')['all_sales'].agg('sum').sort_values(ascending=False).index[0:3])
top5_platform = list(data.groupby('platform')['all_sales'].agg('sum').sort_values(ascending=False).index[0:5])
top10_platform = list(data.groupby('platform')['all_sales'].agg('sum').sort_values(ascending=False).index[0:10])
top20_platform = list(data.groupby('platform')['all_sales'].agg('sum').sort_values(ascending=False).index[0:20])

# создадим вложенный список для цикла
list_platform = [top3_platform, top5_platform, top10_platform, top20_platform]

for platform_name in list_platform:
    # делаем выборку данных для топ-3, топ-5 и т.д. платформ
    temp = data.query('platform in @platform_name')
    
    time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
    time_life['life'] = time_life['max'] - time_life['min'] + 1 # в срок жизни платформы, должны входить границы диапазона, 
                                                            # поэтому +1
    print(f'Медианное время жизни платформы по данным для топ-{len(platform_name)}', time_life['life'].median())
```

**Медианное время жизни платформы уменьшается** с добавлением все менее прибыльных платформ из рейтинга по глобальным продажам, для которых проводится расчет. 
Остановимся на выборке **из топ-10 платформ** для расчета медианного времени, которое составляет **11 лет**. Тогда медианный актуальный период можно рассчитать, как половина от медианного времени жизни, т.е. **5,5 лет**.

### Влияние отзывов пользователей и критиков на продажи игр

Для того, чтобы оценить влияние оценок пользователей и критиков на продажи игр, построим диаграммы рассеяния и рассчитаем корреляцию для этих показателей. 
Так как на этом шаге нам предстоит оценить влияние отзывов пользователей для нескольких игровых платформ, то создадим функцию для отрисовки графиков и вывода коэффициентов корреляции.

```python
# создадим функцию data с 2 параметрами: data - датафрем с данными, platform_name - название платформы, для которой производится расчет
def corr(data, platform_name):
    # cоздадим временный датафрейм со срезом данным по определенной платформе
    data_temp = data_actual[data_actual['platform'] == platform_name] 
    
    # выводи имя платформы
    #print()
   
    # будем отрисовывать 2 графика
    plt.rc('axes', labelsize=8) #fontsize of the x and y labels
    plt.rc('xtick', labelsize=8) #fontsize of the x tick labels
    plt.rc('ytick', labelsize=7) #fontsize of the y tick labels
    fig, axs = plt.subplots(1,2, figsize=(8,3), dpi=100)
    # выводим диаграмму рассеяния для user_score и all_sales
    data_temp.plot(x='user_score', y='all_sales', kind='scatter', 
                   ax=axs[0], xlabel='оценка пользователей', ylabel='глобальные продажи, млн копий',
                   title='Диаграмма рассеяния')
    # выводим диаграмму рассеяния для user_score и all_sales
    data_temp.plot(x='critic_score', y='all_sales', kind='scatter',
                   ax=axs[1], xlabel='оценка критиков', ylabel='глобальные продажи, млн копий', color='darkgreen',
                   title='Диаграмма рассеяния')
    plt.suptitle(f'Платформа {platform_name}', fontsize=10) 
    plt.show()
    #вывовдим коэффициент корреляции
    print(f'Коэффициент корреляции между глобальными продажами игры на платформе {platform_name} и оценкой пользователей ',
          (data_temp['all_sales'].corr(data_temp['user_score'])).round(2))
    #вывовдим коэффициент корреляции
    print(f'Коэффициент корреляции между глобальными продажами игры на платформе {platform_name} и оценкой критиков',
          (data_temp['all_sales'].corr(data_temp['critic_score'])).round(2))
    plt.rcParams.update(plt.rcParamsDefault)
```

Сначала оценим влияние оценок пользователей и критиков на продажи для платформы с наибольшим средним доходом за период 2012-2016 гг. - **PS4**.

```python
corr(data, 'PS4')
```

Как мы видим из графиков и расчетов коэффицентов корреляции, между продажами игр и оценками пользователей **очень слабая отрицательная связь**, между продажами и оценкой критиков - **слабая положительная связь**.

Определим топ-5 самых популярных платформ для пользователей каждого региона. Изобразим результаты в виде круговых диаграмм.

## Проверка гипотез

### Сравнение пользовательские рейтингов платформ Xbox One и PC

Для актуального периода хотим сравнить средние пользовательские рейтинги платформ Xbox One и PC. Таким образом, мы будем сравнивать средние двух независимых выборок.

Сформилируем **нулевую гипотезу**: средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.

**Альтернативная гипотеза двустороняя**: средний пользовательский рейтинг платформы Xbox One не равен среднему пользовательскому рейтингу платформы PC.

Будем использоваться двусторонний t-test для независимых выборок.

```python
#первая выборка - оценки пользователей игр на платформе Xbox One
sample1 = data_actual[(data['platform'] == 'XOne') & (data['user_score'].isna() == False)]['user_score']
#вторая выборка - оценки пользователей игр на платформе PC
sample2 = data_actual[(data['platform'] == 'PC') & (data['user_score'].isna() == False)]['user_score']

# выведем средние значния 2 выборок
print(f'Cредний пользовательский рейтинг платформы Xbox One: {np.mean(sample1)}')
print(f'Средний пользовательский рейтинг платформы PC: {np.mean(sample2)}')

# установим уровень значимости 5 %
alpha = 0.05

#alternative two-sided
results = st.ttest_ind(sample1, sample2)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Гипотезу о равенстве средних пользовательских рейтингов платформы Xbox One и PC отвергнуть не удалось.

### Сравнение пользовательские рейтингов жанров Action и Sports 

**Вопрос**: разные ли средние пользовательские рейтинги жанров Action и Sports?
Для ответа на вопрос будем также использовать  двусторонний t-test для независимых выборок.

**Нулевая гипотеза**: средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.

**Альтернативная гипотеза двусторонняя**: средний пользовательский рейтинг жанра Action не равен среднему пользовательскому рейтингу жанра Sports.

```python
#первая выборка - оценки пользователей игр жанра Action
sample1 = data_actual[(data['genre'] == 'Action') & (data['user_score'].isna() == False)]['user_score']

#вторая выборка - оценки пользователей игр жанра Sports
sample2 = data_actual[(data['genre'] == 'Sports') & (data['user_score'].isna() == False)]['user_score']

# выведем средние значния 2 выборок
print(f'Cредний пользовательский рейтинг жанра Action: {np.mean(sample1)}')
print(f'Средний пользовательский рейтинг жанра Sports: {np.mean(sample2)}')

# установим уровень значимости 5 %
alpha = 0.05

#alternative two-sided
results = st.ttest_ind(sample1, sample2)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

По результатам статистического теста нулевую гипотезу отвергаем в пользу альтернативной о том, что средние пользовательские рейтинги жанров Action и Sports разные.

### Вывод по разделу

После проверки гипотез, мы узнали получили следующие результаты:
* гипотезу **о равенстве** средних пользовательских рейтингов платформы Xbox One и PC **отвергнуть не удалось**.
* гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports **отвергаем** в пользу альтернативной о том, что средние пользовательские рейтинги жанров Action и Sports **разные**.
