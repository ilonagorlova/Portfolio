## Предобработка данных.

### Изменение названия столбца.

Приведем название столбца `cityCenter_nearest` к змеиному регистру и добавим единицы измерения.

```python
# изменим название столбца cityCenters_nearest на cityсenters_nearest_m
data = data.rename(columns={'cityCenters_nearest': 'citycenters_nearest_m'})
```

### Обработка пропущенных значений

```python
# проверим количество пропусков в каждом столбце
data.isna().sum()
```

В большинстве столбцов встречаются пропуски, в отдельных столбцах пропущено более половины значений.
Стоит отметить, что пропуски встречаются в каждом столбце, который заполнялись автоматически на основе картографических данных.

Рассмотрим пропуски в каждом столбце и проанализируем, можно ли их заполнить.

```python
print('Число пропусков в столбце ceiling_height:', data['ceiling_height'].isna().sum())
# посмотрим диапазон значений высоты потолков
print(data['ceiling_height'].sort_values().unique())
```

В столбце с данным о высоте потолков встречаются как **аномально низкие значения меньше 2 м**, так и аномально большие (более 10 м). 
Посмотрим на данные с аномальными значениями высоты потолков.

```python
# выведем срез из датафрейма с квартирами, у которых высота потолка < 2 м
display(data[data['ceiling_height'] < 2])
```

Всего в данных только 3 квартиры, где указаны неправдоподобно низкие высоты потолков. Можно заменить пропущенные и аномально низкие высоты медианным значением (т.к. распределение данных не соответствует нормальному и имеются выбросы). <a id=1></a>

```python
# заменим пропуски и аномально маленькие значения (< 2 м) в столбце ceiling_height медианным значением
data.loc[(data['ceiling_height'].isna()) | (data['ceiling_height'] < 2), 'ceiling_height'] = data['ceiling_height'].median()
```

Далее посмотрим, сколько в данных встречается квартир с **аномально высокими потолками**. Предположим, что данные могли быть указаны в дециметрах. Тогда примем значение 20 (20 дм = 2 м) как пороговое для аномально высоких потолков.

```python
# выведем срез из датафрейма с квартирами, у которых высота потолка >= 20 м
display(data[data['ceiling_height'] >= 20])

# выведем количество строк с ceiling_height >= 20
data[data['ceiling_height'] >= 20].shape[0]
```

В данных встречаются 23 строки с аномально высокими значениями высоты потолков. Если их разделить на 10, то данные выглядят более правдоподобно.

```python
# разделим значения высоты потолков, которые больше или равны 20, на 10
data.update(data[data['ceiling_height'] >= 20]['ceiling_height']/10)
data['ceiling_height'].sort_values().unique()
```

После заполнения отсутствующих значений и замены аномалий в данных о высоте потолков встречаются значения от 2 до 14 м (эти значения тоже кажутся завышенными, вернемся к анализу аномалий далее [в разделе 5.1.5](#2)).

```python
# проверим число пропусков в столбце living_area
print('Число пропусков в столбце living_area:', data['living_area'].isna().sum())
# заполним отсутствующие значения в столбце медианным значениям для выборки данных в зависимости от количества комнат в квартире
for n in data['rooms'].unique():
    data.loc[(data['rooms'] == n) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == n), 'living_area'].median()
# проверим число пропусков в столбце living_area после корректировки
print('Осталось пропусков после корректировки:', data['living_area'].isna().sum())
```

```python
# проверим число пропусков в столбце 'is_apartment'
print('Число пропусков в столбце is_apartment:', data['is_apartment'].isna().sum())
# посчитаем долю пропущенных значений от общего количества строк
data['is_apartment'].isna().sum()/data.shape[0]*100
```

В столбце `is_apartment` очень много пропусков - 20924, что составляет 88 % от всех строк. На то, что квартира относится к аппартаментам, указывает значение `True`. А то, что квартиры не относятся к аппартаментом, могли не обозначить в файле. Так как доля апартаментов на рынке небольшая, предположим, что отсутствующие значения аналогичны значениями `False` и заменим их.

```python
# заменим отсутствующие значения в столбце is_apartment на False
data['is_apartment'] = data['is_apartment'].fillna(False)
# проверим количество пропусков после корректировки
print('Осталось пропусков после корректировки:', data['is_apartment'].isna().sum())
```

```python
# проверим число пропусков в столбце kitchen_area
print('Число пропусков в столбце kitchen_area:', data['kitchen_area'].isna().sum())
# сделаем выборку квартир, у которых отсутствуют данные о площади кухни
data_kitchen_nan = data[data['kitchen_area'].isna()]

# посмотрим, сколько отсутствующих значений площади кухни в группах по количеству комнат
print('Число пропусков площади кухни для каждой группы квартир по количеству комнат')
print(data_kitchen_nan.groupby('rooms')['floor'].count()) 

# посмотрим, сколько всего значений площади кухни в каждой группе по количеству комнат
print('')
print('Число квартир в каждой группе (по количеству комнат), для которых указана площадь кухни')
data.groupby('rooms')['kitchen_area'].count()
```

По приведенным выше данным видно, что пропуски площади кухни встречаются в каждой группе квартир (по количеству комнат). 

Для группа rooms = 0 вообще нет данных о площади кухни во всем датафрейме. Возможно, это связано с тем, что данные относятся к квартирам-студиям, а кухня входит в общую площадь и не указывается отдельно.

Для квартир с количеством комнат от 1 до 9 можно заполнить пропуски медианным значением.

```python
# заполним медианным значением пропуски для групп с количеством комнат (1-9),кроме rooms = 0 (т.к. в этой группе вообще нет данных о площади кухни)
for n in range(1,10):
    data.loc[(data['rooms'] == n) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == n), 'kitchen_area'].median()
print('Осталось пропусков в столбце kitchen_area:', data['kitchen_area'].isna().sum())
```

Проанализируем оставшиеся 197 пропусков площади кухни для группе rooms=0. Проверим, сколько квартир-студий в группе rooms=0.

```python
# проверим, сколько квартир-студий в группе rooms=0
data_kitchen_nan[data_kitchen_nan['rooms'] == 0]['studio'].value_counts()
```

```python
# сделаем выборку данных о квартирах-студиях
data_studio = data[data['studio']]
# посчитаем, сколько квартир студий всего и в группах по количеству комнат (0,1)
print('Всего квартир-студий:', data_studio['floor'].count())
print('Кол-во квартир-студий в группах с разным числом комнат', data_studio['rooms'].value_counts())
```

Как видно из данных, для квартир студий количество комнат может указываться как 0 или как 1 (хотя, по факту, в этих квартирах 1 комната). Всего в данных 149 квартир-студий, их них 138 с количеством комнат = 0 и 11 квартир с 1 комнатой.

На диаграмме размаха определим границы наиболее характерных значений общей площади: примерно **от 10 до 120 м2**.

Чтобы не брать значения на глаз сделаем функцию для расчета граничных значений "усов" диаграммы размаха.


```python
# сделаем функцию для расчета значений "усов" диаграммы размаха (возможно, для этого есть готовый метод, но мы его не проходили пока)
def whiskers_where(column):
    # выводим числовую статистику
    stat = column.describe()
    # рассчитываем значения по формуле
    whisker_max = stat[6] + 1.5*(stat[6] - stat[4])
    whisker_min = stat[4] - 1.5*(stat[6] - stat[4])
    # учитываем минимальное и максимальные значения
    if whisker_min < stat[3]:
        whisker_min = stat[3]
    if whisker_max > stat[7]:
        whisker_max = stat[7]  
    print('Усы диаграммы размаха')
    return whisker_min, whisker_max
```

```python
print(whiskers_where(data['total_area']))
```

На диаграмме размаха усы ограничивают репрезентиативные данные в диапазоне **от 12 до 114.75 м2**, а также наблюдаются выбросы в сторону больших значений.
